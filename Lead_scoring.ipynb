```python
#functions for dummy variable creation
def binary_map(x):
    return x.map({'Yes': 1, "No": 0})
```

```python
def get_categorical_columns(df, unique_val_threshold=0.05):
    """
    Identifies and returns a list of categorical columns in a DataFrame.

    Returns:
    - A list of column names that are considered categorical.
    """
    categorical_cols = []
    for col in df.columns:
        if pd.api.types.is_categorical_dtype(df[col]):
            categorical_cols.append(col)
        else:
            unique_count = df[col].nunique()
            total_count = len(df[col])
            if unique_count / total_count < unique_val_threshold:
              categorical_cols.append(col)
    return categorical_cols

categorical_vars = get_categorical_columns(leads_data)
```

```python
def identify_and_convert_yes_no_columns(df, columns):
    """
    Identifies columns with only 'Yes' or 'No' values and applies a binary mapping.

    Returns:
    - A list of columns that were identified as having only 'Yes'/'No' values
      and were converted to binary format.
    """
    converted_columns = []
    for col in columns:
        if set(df[col].dropna().unique()).issubset({'Yes', 'No'}):
            df[col] = binary_map(df[col])
            converted_columns.append(col)
    return converted_columns
```

```python
from sklearn.model_selection import train_test_split
# Creating X (Independent variables) dataframe
X = leads_data.drop(['Converted'], axis=1)
X.head()
```

```python
# Creating y (Target variable) dataframe
y = leads_data['Converted']
y.head()
```

```python
# Now we will perform test train sptil
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
# After train test split, We will do scaling using fit_transform function
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']] = scaler.fit_transform(X_train[['TotalVisits','Total Time Spent on Website','Page Views Per Visit']])

X_train.head()
```

```python
Converted = (sum(leads_data['Converted'])/len(leads_data['Converted'].index))*100
Converted
```

```python
# Feature selection
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()

from sklearn.feature_selection import RFE
rfe = RFE(estimator=logreg, n_features_to_select=20)
rfe = rfe.fit(X_train, y_train)
```

```python
# Selected features
rfe.support_
```

```python
# Listing selected features rank wise
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
cols = X_train.columns[rfe.support_]
cols
```

```python
import statsmodels.api as sm
from sklearn.metrics import r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Calculation of VIF
def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif
```

```python
# Creating X train with RFE Selected Variables
X_train_rfe = X_train[cols]
# Adding Constant
X_train_rfe = sm.add_constant(X_train_rfe)
calculateVIF(X_train_rfe)
```

```python
# Adding Constant
X_train_lm_1 = sm.add_constant(X_train_rfe)

# Create Object for the Model
lr_model = sm.OLS(y_train,X_train_lm_1,family = sm.families.Binomial()).fit()

# View Summary of Statistics
print(lr_model.summary())
```

```python
# As VIF is high for the Feature 'Lead Origin_Lead Add Form' so we can drop it
X_train_rfe = X_train_rfe.drop(['Lead Origin_Lead Add Form'], axis = 1)

calculateVIF(X_train_rfe)
```

```python
# Model 2
X_train_lm_2 = sm.add_constant(X_train_rfe)
lr_model = sm.OLS(y_train,X_train_lm_2,family = sm.families.Binomial()).fit()
print(lr_model.summary())
```

```python
X_train_rfe = X_train_rfe.drop(['Last Notable Activity_Had a Phone Conversation'], axis = 1)

calculateVIF(X_train_rfe)
```

```python
#model 3
X_train_lm_3 = sm.add_constant(X_train_rfe)
lr_model = sm.OLS(y_train,X_train_lm_3,family = sm.families.Binomial()).fit()
print(lr_model.summary())
```

```python
X_train_rfe = X_train_rfe.drop(['Last Activity_Email Bounced'], axis = 1)

calculateVIF(X_train_rfe)
```

```python
# Model 4
X_train_lm_4 = sm.add_constant(X_train_rfe)
lr_model = sm.OLS(y_train,X_train_lm_4,family = sm.families.Binomial()).fit()
print(lr_model.summary())
```

```python

y_train_pred = lr_model.predict(X_train_lm_4)
y_train_pred[:10]
```
